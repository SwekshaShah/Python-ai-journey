pop(): it remove the last element

```python
cart=["shoes","belt","shirt","hat"]
cart.pop()
print(cart)
```

pop(indexing): it will remove the elment of the given indexing

```python
cart=["shoes","belt","shirt","hat"]
cart.pop(2)
print(cart)
```

len(): to count the items

```python
cart=["shoes","belt","shirt","hat"]
print(len(cart))
```

```python
cart=["shoes","belt","shirt","hat"]
cart.append("pants")
cart.insert(1,"shoes")
print(cart.index("hat"))
print(cart.count("shoes"))
```

list_name.copy(): to copy the list

```python
cart=["shoes","belt","shirt","hat"]
cart2=cart.copy()
print(cart)
print(cart2)
```
