```python
def print_all(*args):
    for arg in args:
        print(arg)

print_all("hello","world",1,2,3)
```

**kwargs(variable keyword):
used toaccept number of keyword(named) argument
it is packed in dictionary


```python
def print_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
    print(kwargs)
print_info(name="sweksha",age=18,city="delhi")
```

```python
def create_user(username,**profile):
    return {"username":username,
            "profile":profile}

user=create_user("sweksha",age=18,city="delhi")
print(user)
```

```python
def print_all(a,b,*args,**kwargs):
    print(a+b+sum(args)+sum(kwargs.values()))
print_all(1,2,3,4,5,x=10,y=20)
```

```python
def get_stats(numbers):
    return min(numbers),max(numbers),sum(numbers)/len(numbers)
min_value,max_value,average=get_stats([1,2,3,4,5])
print(get_stats([1,2,3,4,5]))
print(f"Min:{min_value},Max:{max_value},Average:{average}")
```

```python
numbers=[3,1,4,1,5,9,2,6]
sorted_numbers=sorted(numbers,key=lambda x:abs(x-5))
print(sorted_numbers)
```

```python
numbers=[3,1,4,1,5,9,2,6]
evens=list(filter(lambda x:x%2==0,numbers))
print(evens)
```

```python
numbers=[3,1,4,1,5,9,2,6]
doubled=list(map(lambda x:x*2,numbers))
print(doubled)
```

```python
numbers=[3,1,4,1,5,9,2,6]
mapped=list(map(lambda x:x*2,numbers))
print(mapped)
mapped1=[x*2 for x in numbers]
print(mapped1)
```
