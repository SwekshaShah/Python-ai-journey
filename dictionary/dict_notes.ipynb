Dictionary:A dictionary is maps keys to values like a real dictionary maps words to definition.
keys are how you look things up- they must be unique and immutuable(string, ints,tuples). Values can be anything.

```python
person={
    "name":"sweksha",
    "age":18,
    "city":"kathmandu"
}
print(person['name'])
print(person['age'])
print(person['city'])
```

```python
person={
    "name":"sweksha",
    "age":18,
    "city":"kathmandu"
}
person['email']="sweksha234@gmail.com"
person['age']=20
print(person)
```

update:.update()

```python
user={'name':"sweksha",'age':18}
user.update({'email':"swekshs123@gmail.com",'city':"ktm"})
print(user)
```

.keys():use to pull keys

```python
user={'name': 'sweksha', 'age': 18, 'email': 'swekshs123@gmail.com', 'city': 'ktm'}
print(user.keys())
print(list(user.keys()))
```

```python
user={'name': 'sweksha', 'age': 18, 'email': 'swekshs123@gmail.com', 'city': 'ktm'}
print(user.values())
print(list(user.values()))
```

```python
user={'name': 'sweksha', 'age': 18, 'email': 'swekshs123@gmail.com', 'city': 'ktm'}
print(user.items())
print(list(user.items()))
```

```python
user={'name': 'sweksha', 'age': 18, 'email': 'swekshs123@gmail.com', 'city': 'ktm'}
for key in user:
    print(key,":",user[key])

for key,value in user.items(): #another way to write same thing 
    print(f"{key}:{value}")
```

```python
user={'name': 'sweksha', 'age': 18, 'email': 'swekshs123@gmail.com', 'city': 'ktm'}
user.pop("age")
user.pop("number","not available") #best way by adding default
print(user)
```

```python
user={'name': 'sweksha','email': 'swekshs123@gmail.com', 'city': 'ktm'}
if "name" in user:
    print("exist")
if "age" in user:
    print("don't exist")
```

```python
items=["apple","banana","orange","mango"]
prices={item:(i+1)*10 for i, item in enumerate(items)}
print(prices)
```

```python
data={"a":1,"b":2,"c":3}
flipped={v:k for k,v in data.items()} #swap key and value
print(flipped)
```

```python
transactions=[
    {"amount":100,"category":"food"},
    {"amount":200,"category":"transporation"},
    {"amount":3000,"category":"shopping"}
]
categories=set(t["category"] for t in transactions)
by_category= {
    cat:[t["amount"] for t in transactions if t["category"]==cat]
    for cat in categories
}
#{"food":[100],"transportation":[200],"shopping":[3000]}
print(by_category)
```
