lambda function:
it is a small anonymous(has no name)function in python.


```python
def square(x):
    return x*x
print(square(5))
```

using lambda
(lambda parameter:expression)

```python
square=lambda x:x*x
print(square(5))
```

using lambda with map()
map(lambda parameter: expression, iterable)

```python
number=[1,2,3,4,5]
square=list(map(lambda x:x*x,number))
print(square)
```

sorted()
sorted() returns a new sorted list from any iterable (list, tuple, set, etc.)
sorted(iterable, key=function, reverse=True/False)

```python
word=["apple","kiwi","banana","grape"]
result=sorted(word,key=lambda x:len(x))
print(result)
```

filter():keeps selected items

```python
number=[1,2,3,4,5]
result=list(filter(lambda x:x%2==0,number))
print(result)
```

enumerate(): used when you want to get both index and value while looping
syntax:enumerate(iterable,start=0)

```python
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits):
    print(index, value)
```

```python
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits, start=1):
    print(index, value)
```

```python
cart=[]
def show_cart():
    if not cart:   # or if cart==[]:
        print("Your cart is empty:")
        return
    print("Your cart")
    total=0
    for i, item in enumerate(cart):
        print(f"{i+1}.{item["name"]:<15}{item["price"]:>6.2f}")
        total+=item['price']
    print()
    print(f"{"total":<15} {total:>6.2f}")
```

```python
cart = [
    {"name": "bag", "price": 1000},
    {"name": "book", "price": 3000}
]

def show_cart():
    if not cart:
        print("Your cart is empty:")
        return

    print("Your cart")
    total = 0

    for i in range(len(cart)):
        item = cart[i]
        print(f"{i+1}. {item['name']:<15}{item['price']:>6.2f}")
        total += item['price']

    print()
    print(f"{'total':<15} {total:>6.2f}")

show_cart()
```

```python
cart = [
    {"name": "bag", "price": 1000},
    {"name": "book", "price": 3000},
    {"name": "shoes", "price": 5000}
]

def search_cart():
    query = input("Search item name: ").lower()

    results = [item for item in cart if query in item['name'].lower()]

    if results:
        print("Search results:")
        for item in results:
            print(f"{item['name']:<15}{item['price']:<6.2f}")
    else:
        print(f"No '{query}' found in cart")

search_cart()
```
