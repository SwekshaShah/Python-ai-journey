you have alist student test scores. You need to create a new list called a new list called grades that converts those numbers into strings: "Pass" if the score is 20 or higher and"fail" if its below 20.

```python
score= [25,15,24,45,32,1]
grade=[]
for i in score:
  grade.append("pass") if i>=20 else grade.append("fail")
print(grade)
```

another way to solve the same question.

```python
score= [25,15,24,45,32,1]
grades=["Pass" if i>=20 else "Fail" for i in score]
print(grades)
```
